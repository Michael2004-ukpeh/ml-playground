# Utility Functions


```python
# Remove references helper
def remove_refs(x):
    if pd.isna(x):
        return np.nan
    return re.sub(r"\[.*?\]", "", str(x))
```

# Clean Peak

```python
df["Peak"]
```

```python
df["Peak"] = df["Peak"].apply(remove_refs).astype(float)
df["Peak"]
```

# Clean All time peak

```python
df["All Time Peak"]= df['All Time Peak'].apply(remove_refs).astype(float)
df["All Time Peak"]
```

# Clean Currency related Fields

```python
# Clean currency columns to remove dollar sign and commas
# \u00A is non-breaking space in pandas
currency_cols = [
    "Actual\u00A0gross",
    "Adjusted\u00A0gross (in 2022 dollars)",
    "Average gross"
]

for col in currency_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(r"\[.*?\]", "", regex=True)
        .str.replace("$", "", regex=False)
        .str.replace(",", "", regex=False)
        .astype(float)
    )

df[currency_cols]
```

# Clean Text in Artist and Tour Title columns


```python
df["Artist"] = df["Artist"].apply(remove_refs)

df["Tour title"] = (
    df["Tour title"]
    .astype(str)
    .str.replace(r"\[.*?\]", "", regex=True)
    .str.replace("†", "", regex=False)
    .str.replace("‡", "", regex=False)
    .str.replace("*", "", regex=False)
    .str.strip()
)

df[["Artist", "Tour title"]]
```

# Clean Years column
By engineering, start date and end date by splitting the years




```python
df["Year(s)"]
# Feature engineer Year(s)
def split_years(x):
    if pd.isna(x):
        return pd.Series([np.nan, np.nan, np.nan])
    years = re.split("–|-", x)
    start = int(years[0])
    end = int(years[-1])
    return pd.Series([start, end, end - start + 1])

df[["start_year", "end_year", "tour_duration_years"]] = (
    df["Year(s)"].apply(split_years)
)
df[["start_year", "end_year", "tour_duration_years"]]
```

# Make sure show columns are integers

```python
df["Shows"] = df["Shows"].astype(int)
```

# Remove ref and year(s) columns

```python
df.drop(columns=[ "Ref.", "Year(s)"], inplace=True)
```

# Clean Rank column, eliminating duplicate ranks by ranking revenue

```python
df = df.sort_values(
    by="Adjusted\u00A0gross (in 2022 dollars)",
    ascending=False
)

df["Rank"] = range(1, len(df) + 1)
```

## Implement Missing Values with Pandas


Use pandas to impute `NaN` values in the 'Peak' and 'All Time Peak' columns with their respective median values. Display `df.isnull().sum()` again to confirm no NaNs are left in these columns, and then display `df.head()` to show the cleaned DataFrame.

# Implement Feature Engineering with Scikit-learn Pipelines

Identify numerical columns that are suitable for scaling and categorical columns for one-hot encoding. Create a `ColumnTransformer` to apply `StandardScaler` to numerical features ('Peak', 'All Time Peak', 'Shows', 'Average gross', 'start_year', 'end_year', 'tour_duration_years') and `OneHotEncoder` to the 'Artist' and 'Tour title' columns. Then, fit and transform the DataFrame to produce the final feature-engineered dataset.


```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import pandas as pd

numerical_cols = [
    'Rank',
    'Peak',
    'All Time Peak',
    'Shows',
    'Average gross',
    'start_year',
    'end_year',
    'tour_duration_years',
    'Actual\u00A0gross',
    'Adjusted\u00A0gross (in 2022 dollars)'
]
categorical_cols = ['Artist', 'Tour title']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ])

df_processed = preprocessor.fit_transform(df)
feature_names = preprocessor.get_feature_names_out()

df_final = pd.DataFrame(df_processed, columns=feature_names)

print(df_final.info())
df_final
```
