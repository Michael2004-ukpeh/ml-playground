## Normalize given independent data


```python
x1 = df['x1'].values
x2 = df['x2'].values
x3 = df['x3'].values
Y = df['label'].values

X1 = (x1 - np.min(x1))/(np.max(x1)-np.min(x1))
X2 = (x2 - np.min(x2))/(np.max(x2)-np.min(x2))
X3=  (x3 - np.min(x3))/(np.max(x3)-np.min(x3))
display(X1,X2,X3)
```

# Utility and Training functions

```python
# Loss Function
def binary_cr_entropy(Y, Y_pred):
  return -np.mean(Y*np.log(Y_pred) + (1-Y)*np.log(1-Y_pred))
```

```python
# Probability function
def sigmoid(z):
  # Where z = m1*X1 + m2*X2 + m3*X3 + c
  return 1/(1+ np.exp(-z))
```

```python
# Gradient descent
def gradient_change(m1, m2, m3, c, X1, X2, X3, Y):
  """
  X1 is the independent variable
  X2 is the independent variable
  X3 is the independent variable
  Y is the dependent variable
  m1 is the current value of m1
  m2 is the current value of m2
  m3 is the current value of m3

  returns the gradient of m1, m2 and m3 and intercept
  """
  Y_pred = sigmoid(m1*X1 + m2*X2 + m3*X3 + c)

  # Calculate error/raw residual
  error = Y_pred - Y

  # Calculate gradient
  dc = np.mean(error)
  dm1 = np.mean(error*X1)
  dm2 = np.mean(error*X2)
  dm3 = np.mean(error*X3)

  return dm1, dm2, dm3, dc
```

# Let's train
- Calculate prediction based on sigmoid function
- Calculate cost based on binary cross entropy
- Calculate gradients of parameters
- Finetune parameters using gradient decent




```python
def train(X1, X2, X3, Y, m1,m2,m3, c, lr, epochs, history):
  for i in range(epochs):
    # compute current loss function
    Y_pred = sigmoid(m1*X1 + m2*X2 + m3*X3 + c)
    cost= binary_cr_entropy(Y, Y_pred)
    dm1, dm2, dm3, dc = gradient_change(m1, m2, m3, c,X1, X2, X3, Y)

    # Track hyperparameters
    history['m1_arr'].append(m1)
    history['m2_arr'].append(m2)
    history['m3_arr'].append(m3)
    history['c_arr'].append(c)
    history['err_arr'].append(cost)

    # Update hyperparameters
    m1 = m1 - lr*dm1
    m2 = m2 - lr*dm2
    m3 = m3 - lr*dm3
    c = c - lr*dc

    print(f"Epoch {i} :Binary Cross Entropy = {cost}\n ")
  print(f"Final parameters: m1 = {m1}, m2 = {m2}, m3 = {m3}, c = {c}")
  return m1, m2, m3, c
```

Train with initial values of coefficients

```python
m1, m2, m3, c = 0.5, 0.5, 0.5, 0.5
lr = 0.1
epochs = 500000

history = {
    'm1_arr': [],
    'm2_arr': [],
    'm3_arr': [],
    'c_arr': [],
    'err_arr': []
}

m1, m2, m3, c = train(X1,X2,X3,Y,m1,m2,m3,c,lr,epochs,history)
```

# Graph of Cost function as training progresses

```python
plt.figure(figsize=(8, 6))
plt.plot(range(1,epochs + 1 ),history['err_arr'])

plt.xlabel("Number of Epochs")
plt.ylabel("Binary Cross Entropy")
plt.title("Binary Cross Entropy vs Number of Epochs")
plt.grid(True)
plt.legend()

plt.show()
```
